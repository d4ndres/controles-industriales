# 1.- Importar librerias

matplotlib es una libreria de python que permite realizar diferentes tipos de gráficas de manera muy sencilla, incluyendo de dos y tres dimensiones

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

# 2.- Gráficas en dos dimensiones

```python
#Crear el vecto de tiempo
x = np.linspace(0,2*np.pi,200)

#Creamos la funcion
y = 2 * np.cos(x)

#Creamos la gráfica
plt.figure(1)
plt.title("Función 2*cos(x)")
plt.plot(x,y, color="green")
plt.xlabel("Tiempo(x)")
plt.ylabel("Señal de salida")
plt.ylim([-2.1,2.1])
plt.xlim([-0.1, 2*np.pi+0.1])
plt.grid()
plt.show()
```

```python
#Generamos la función seno
y2 = 2 * np.sin(x)

#Generamos la gráfica compuesta
plt.figure(2)
plt.title("Funciones Trigonométricas")
ax, = plt.plot(x, y, "r", label="cos(x)")
bx, = plt.plot(x, y2, "b", label="sin(x)")

#Insertamos la leyenda y las etiquetas
plt.legend(handles=[ax, bx])
plt.xlabel("Tiempo")
plt.ylabel("Señal de salida")
plt.ylim([-2.1,2.1])
plt.xlim([-0.1, 2*np.pi+0.1])
plt.grid()
plt.show()
```

```python
#Uso del subplot
plt.figure(3)
plt.subplot(2,1,1)
plt.plot(x,y, c='y')
plt.ylabel("cos(x)")
plt.grid()
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.ylabel("sin(x)")
plt.xlabel("Tiempo")
plt.grid()
plt.show()
```

# 3.- Graficas tipo scatter

Las gráficas de tipo scatter permiten representar puntos de dos dimensiones que poseen algún tipo de clasificación

```python
#Generar el conjunto de puntos
x = np.zeros([200,2])
x[:,0] = 0.4+0.2*np.random.rand(200)
x[:,1] = 0.4-0.2*np.random.rand(200)
y = np.zeros(200)


#Generar el arreglo
plt.figure(4, figsize=(4,4))
plt.scatter(x[:,0], x[:,1], c=y, s=40, cmap=plt.cm.Spectral)
```

```python
x = np.zeros([400,2])
y = np.zeros(400)

#formar el cluster 1
x[0:200,0] = 0.4+0.2*np.random.rand(200) - 0.1*np.random.rand(200)
x[0:200,1] = 0.4+0.2*np.random.rand(200) - 0.1*np.random.rand(200)
y[0:200] = np.zeros(200)

#formar el cluster 2
x[200:400,0] = 0.6+0.2*np.random.rand(200) - 0.1*np.random.rand(200)
x[200:400,1] = 0.6+0.2*np.random.rand(200) - 0.1*np.random.rand(200)
y[200:400] = np.ones(200)

#Generar el arreglo
plt.figure(5, figsize=(4,4))
plt.scatter(x[:,0], x[:,1], c=y, s=40, cmap=plt.cm.Spectral)
# plt.scatter(x[0:200,0], x[0:200,1], c='#FFCE3C', s=40, cmap=plt.cm.Spectral)
# plt.scatter(x[200:400,0], x[200:400,1], c='purple', s=40, cmap=plt.cm.Spectral)
plt.grid()
plt.show()
```

La clase mpl_toolkits provee herramientas para representar puntos tridimensionales de manera sencilla

```python
#Indicamos que la importe en ete proyecto
# %matplotlib notebook

#Creamos la rejilla

x = np.zeros([400,2])
for i in range(20):
    x[i*20:(i+1)*20,0] = i/20
    x[i*20:(i+1)*20,1] = np.linspace(0,100,20) / 100

#Creamos la función 3d que vamos a gráfica
y = np.power(x[:,0] - 0.5, 3) + np.power(x[:,1] - 0.5, 3)

#Construimos la gráfica
plt.figure(6)
ax = plt.axes(projection='3d')
ax.scatter(x[:,0], x[:,1], y, marker='*')
plt.show()

    
```
